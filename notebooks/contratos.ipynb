```python
import requests
import json
```

```python
content_data = requests.get('http://127.0.0.1:5000/api/contratos/dependencia-cantidad')
response = json.loads(content_data.content)
```

```python
import matplotlib.pyplot as plt
import numpy as np
```

```python
# make data:
names = [d['nombre_dependencia'][0:3] for d in response]
values = [d['count'] for d in response]

# plot
fig, ax = plt.subplots()

ax.bar(names, values, width=1, edgecolor="white", linewidth=0.7)

ax.set(xlim=(0, len(response)), xticks=np.arange(0, len(response)),
    ylim=(0, 200), yticks=np.arange(1, 200))

plt.show()
```

```python
plt.style.use('_mpl-gallery-nogrid')


# make data
x = [d['count'] for d in response]
y = [d['nombre_dependencia'][0:5] for d in response]
colors = plt.get_cmap('Blues')(np.linspace(0.2, 0.7, len(x)))

# plot
fig, ax = plt.subplots()
ax.pie(x,labels=y, colors=colors, radius=10, center=(4, 4),
    wedgeprops={"linewidth": 1, "edgecolor": "white"}, frame=True)

# ax.set(xlim=(0, 50), xticks=np.arange(1, 50),
#     ylim=(0, 50), yticks=np.arange(1, 50))

plt.show()
```
